```python
def f(x):
    return (x + 1) / (x - 2)

def p(f, a, b, sh):
    x = np.linspace(a, b - sh, int((b - a) / sh))
    y = sh * f(x) 
    return np.sum(y)

def t(f, a, b, sh):
    x = np.linspace(a, b, int((b - a) / sh) + 1)
    y = f(x[0]) + 2 * np.sum(f(x[1:-1])) + f(x[-1])
    return y * sh / 2

def s(f, a, b, sh):
    x = np.linspace(a, b, int((b - a) / sh) + 1)
    y = f(x[0]) + 4 * np.sum(f(x[1:-1:2])) + 2 * np.sum(f(x[2:-1:2])) + f(x[-1])
    return y * sh / 3
```
