3. В Scipy также есть много менее распространенных специальных функций. Взгляните на обширный список <a href=https://docs.scipy.org/doc/scipy/reference/special.html> здесь</a>!
```python
x = np.linspace(-1, 1,250)
y = x**2*np.sin(1/x**2)+x
plt.plot(x, y, color = 'violet')
plt.xlabel('Ось x')
plt.ylabel('Ось y')
```

```python
x = np.linspace(5,25)
y = abs(1-((1/(1+x**2))/(1/x**2)))

fig, ax = plt.subplots()
ax.semilogy(x, y, color = 'silver')
ax.set_xlabel('Ось x')
ax.set_ylabel('Ось y')
ax.set_title('график')
```

```python
#gamma
x = np.linspace(-3.5, 5.5, 2251)
y = sc.gamma(x)

plt.plot(x, y, 'b', alpha=0.6, color = 'red')
k = np.arange(1, 7)
plt.plot(k, sc.factorial(k-1), 'k*', alpha=0.6,
label='(x-1)!, x = 1, 2, ...')
plt.xlim(-3.5, 5.5)
plt.ylim(-10, 25)
plt.grid()
plt.xlabel('x')
plt.show()
```

```python
#diric
x = np.linspace(-8*np.pi, 8*np.pi, num=201)
plt.figure(figsize=(8, 8));
for idx, n in enumerate([2, 3, 4, 9]):
    plt.subplot(2, 2, idx+1)
    plt.plot(x, sc.diric(x, n), color = 'orange')
    plt.title('diric, n={}'.format(n))
plt.show()
```

## Exercise 2  
1. Построить график функции $sin(x)/x$ на отрезке $[-6\pi;6\pi]$ и добавить тики по оси $x$ с шагом $\pi/2$ от $-6\pi$ до $6\pi$, причём обозначения тиков должны содержать греческую букву $\pi$. Добавить сетку на график.

```python
x = np.linspace(-6*np.pi, 6*np.pi,24)
y = np.sin(x)/x
fig, ax = plt.subplots(figsize=(16,16))
ax.plot(x, y, color = 'gold')
plt.xticks(
    x,
    [r'$-6\pi$', r'$-5.5\pi$', r'$-5\pi$', r'$-4.5\pi$', r'$-3.5\pi$', r'$-3\pi$', r'$-2.5\pi$',
    r'$-2\pi$', r'$-1.5\pi$', r'$-\pi$', r'$-0.5\pi$', r'0', r'$0.5\pi$', r'$1\pi$', r'$1.5\pi$', r'$2\pi$',
    r'$2.5\pi$', r'$3\pi$', r'$3.5\pi$', r'$4\pi$', r'$4.5\pi$', r'$5\pi$', r'$5.5\pi$', r'$6\pi$',])
ax.set_xlabel('Ось x')
ax.set_ylabel('Ось y')
ax.grid()
```

2. Создайте цветовой график $e^{-x^2-y^2}$ на интервале $[−3,3]×[−3,3]$.

```python
x = y = np.linspace(-np.pi, np.pi, 100)
X, Y = np.meshgrid(x, y)
fig, axs = plt.subplots()
axs.contourf(X , Y, np.sin(X) * np.sin(Y))
```
